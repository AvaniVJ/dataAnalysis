```python
df=pd.read_csv("pizza_Sales.csv")
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y')
```

```python
total_revenue = df['total_price'].sum()
total_orders = df['order_id'].nunique()
total_pizzas_sold = df['quantity'].sum()
average_order_value = total_revenue / total_orders
average_pizzas_per_order = total_pizzas_sold / total_orders
```

```python
kpi_results = {
    "Total Revenue": total_revenue,
    "Total Orders": total_orders,
    "Total Pizzas Sold": total_pizzas_sold,
    "Average Order Value": average_order_value,
    "Average Pizzas per Order": average_pizzas_per_order,
}
```

```python
kpi_results
```

```python
# Extract day names from order_date
df['day_of_week'] = df['order_date'].dt.day_name()

# Group by day of the week and get total orders
weekly_day_orders = df.groupby('day_of_week')['order_id'].nunique()

# Reorder days of the week correctly
order_of_days = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
weekly_day_orders = weekly_day_orders.reindex(order_of_days)

# Plot bar chart for orders by day of the week
plt.figure(figsize=(10, 5))
sns.barplot(x=weekly_day_orders.index, y=weekly_day_orders.values, color='skyblue')

plt.xlabel("Day of the Week")
plt.ylabel("Total Orders")
plt.title("Total Orders by Day of the Week")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
# Extract month names from order_date
df['month_name'] = df['order_date'].dt.month_name()

# Group by month and get total orders
monthly_orders_bar = df.groupby('month_name')['order_id'].nunique()

# Define correct order of months
order_of_months = ["January", "February", "March", "April", "May", "June", 
                    "July", "August", "September", "October", "November", "December"]
monthly_orders_bar = monthly_orders_bar.reindex(order_of_months)
```

```python
# Plot monthly trend for total orders using a line chart
plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_orders_bar.index, y=monthly_orders_bar.values, marker="o", color='blue', linewidth=2)

plt.xlabel("Month")
plt.ylabel("Total Orders")
plt.title("Monthly Trend of Total Orders")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
category_sales = df.groupby('pizza_category')['total_price'].sum()

# Plot pie chart for sales by pizza category
plt.figure(figsize=(8, 8))
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Percentage of Sales by Pizza Category")
plt.show()
```

```python
# Aggregate total sales by pizza size
size_sales = df.groupby('pizza_size')['total_price'].sum()

# Plot pie chart for sales by pizza size
plt.figure(figsize=(8, 8))
plt.pie(size_sales, labels=size_sales.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Percentage of Sales by Pizza Size")
plt.show()
```

```python
# Create a function for a donut chart
def plot_donut_chart(data, title, colors):
    plt.figure(figsize=(8, 8))
    wedges, texts, autotexts = plt.pie(
        data, labels=data.index, autopct='%1.1f%%', colors=colors,
        wedgeprops=dict(width=0.3)  # Creates the donut effect
    )
    plt.setp(autotexts, size=10, weight="bold")
    plt.title(title)
    plt.show()

# Aggregate sales by pizza category and pizza size
category_sales = df.groupby('pizza_category')['total_price'].sum()
size_sales = df.groupby('pizza_size')['total_price'].sum()

# Define color palettes
category_colors = sns.color_palette("pastel", len(category_sales))
size_colors = sns.color_palette("pastel", len(size_sales))

# Plot donut charts
plot_donut_chart(category_sales, "Percentage of Sales by Pizza Category", category_colors)
plot_donut_chart(size_sales, "Percentage of Sales by Pizza Size", size_colors)
```

```python
# Aggregate total pizzas sold by category
category_pizzas_sold = df.groupby('pizza_category')['quantity'].sum().sort_values(ascending=False)
```

```python
# Plot horizontal bar chart for total pizzas sold by category
plt.figure(figsize=(10, 6))
sns.barplot(y=category_pizzas_sold.index, x=category_pizzas_sold.values, palette="Blues_r")

plt.ylabel("Pizza Category")
plt.xlabel("Total Pizzas Sold")
plt.title("Total Pizzas Sold by Pizza Category")
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Annotate values on bars
for i, value in enumerate(category_pizzas_sold.values):
    plt.text(value + 200, i, str(value), va='center', fontsize=12, fontweight='bold')

plt.show()
```

```python
# Aggregate data for best-selling pizzas
top_pizzas_revenue = df.groupby('pizza_name')['total_price'].sum().nlargest(5)
top_pizzas_quantity = df.groupby('pizza_name')['quantity'].sum().nlargest(5)
top_pizzas_orders = df.groupby('pizza_name')['order_id'].nunique().nlargest(5)

# Plot best-selling pizzas
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Revenue
sns.barplot(
    y=top_pizzas_revenue.index,
    x=top_pizzas_revenue.values,
    hue=top_pizzas_revenue.index,
    palette="Greens_r",
    ax=axes[0],
    legend=False
)
axes[0].set_title("Top 5 Pizzas by Revenue")
axes[0].set_xlabel("Total Revenue")

# Quantity
sns.barplot(
    y=top_pizzas_quantity.index,
    x=top_pizzas_quantity.values,
    hue=top_pizzas_quantity.index,
    palette="Blues_r",
    ax=axes[1],
    legend=False
)
axes[1].set_title("Top 5 Pizzas by Quantity Sold")
axes[1].set_xlabel("Total Quantity Sold")

# Orders
sns.barplot(
    y=top_pizzas_orders.index,
    x=top_pizzas_orders.values,
    hue=top_pizzas_orders.index,
    palette="Oranges_r",
    ax=axes[2],
    legend=False
)
axes[2].set_title("Top 5 Pizzas by Total Orders")
axes[2].set_xlabel("Total Orders")

plt.tight_layout()
plt.show()
```
